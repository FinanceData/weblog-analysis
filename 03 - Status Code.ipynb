## DataFrame.resample()

```python
t = df[df['status'] == 200]['status'].resample('5T', how='count')
```

```python
t.plot()
```

여러 Series를 생성하고, DataFrame으로 조합

```python
status_codes = [200, 201, 206, 301, 302, 307, 400, 403, 500, 503, 900, 904]

dur = '15T'
ser_list = {}

for i in status_codes:
    ser = df[df['status'] == i]['status'].resample(dur, how='count')
    ser_list[i] = ser.fillna(0) 
```

```python
df_status = DataFrame(ser_list)
df_status.head()
```

```python
df_status.plot(secondary_y=[200], figsize=(16,6))
```

```python
# 2XX을 제외
df_status2 = df_status[[301, 302, 307, 400, 403, 500, 503, 900, 904]]
```

```python
df_status2.plot(kind='barh', stacked=True, figsize=(16,20))
```
