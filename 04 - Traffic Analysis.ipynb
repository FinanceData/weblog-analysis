## 트래픽 plot

```python
import pandas as pd
import numpy as np
from pandas import DataFrame

df = pd.read_pickle('data/web_20140408.pkl')
```

```python
df['len'].plot()
```

## resample and  plot (1)
* 10분 단위 요청수(count)합

```python
df_cnt = df['ip'].resample('10T', how='count')
```

```python
df_cnt.head(10)
```

```python
df_cnt.plot(title='Request Count (IP)')
```

## resample and  plot (2)
* 5분 단위 트래픽(sum)을 합산
* count(요청수), sum(트래픽)을 함계 plot(secondary_y)

```python
df_len = df['len'].resample('5t', how=['count', 'sum'])
```

```python
df_len.plot(secondary_y=['sum'])
```

```python
df_len.head()
```

## 상관관계
요청수(count)와 트래픽(sum)의 상관계수(Correlation coefficient)


```python
df_len.corr()
```

## 시간 범위 지정
시간의 범위를 지정(07:00 ~ 09:30) 1분 단위, 요청수(count)

```python
df_peak = df['2014-04-08 07:00':'2014-04-08 09:30']['len'].resample('1T', how=['count'])
```

```python
df_peak.plot()
```

## Filtering
2MB 이상의 다운로드 트래픽

```python
df_t = df[df['len'] > 2048 * 1024]
print len(df_t)
```

## 히스토그램 (1)
* 4K 미만의 요청들을 추출
* 0~4K를 100개 구간으로 나누어 히스토그램

```python
df_sm = df[df['len']<4096]['len']
print len(df_sm)
```

```python
df_sm.hist(bins=100)
```

## 히스토그램(2)
4K~8K 트래픽에 대한 히스토그램(100개 구간)

```python
# distribution 4K~8K 
df_sm2 = df[ (4096<=df['len']) & (df['len']<=8192) ]
df_sm2['len'].hist(bins=100)
```
